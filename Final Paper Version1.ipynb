## Introduction
In an era where travel has become a cornerstone of global connectivity, understanding the intricacies of transportation and hospitality industries is more critical than ever. This project explores two creative and impactful questions that analyze key challenges faced by airlines and hotels, aiming to uncover actionable insights for both industry stakeholders and consumers. By exploring data on flight schedules and hotel bookings, we find two most useful questions: How do airlines compare on scheduling accuracy? and What is the association between lead time and hotel cancellations? These questions are motivated by the desire to improve reliability for travelers and efficiency for businesses.

The first question focuses on flight delays, a common issue for travelers. Delays don’t just cause inconvenience; they also create ripple effects, from missing connecting flights to arriving late for important events. Our goal it to highlight which ones are the most reliable by comparing how different airlines handle their schedules. This analysis can help travelers choose airlines that are more likely to get them to their destinations on time and encourage airlines to improve their operations, be more competitive. Our motivation is to help reduce the unpredictability of air travel through finding the most reliable airlines and identifying patterns in delays.

The second question looks at hotel cancellations, a problem that affects both hotels and their guests. When a guest cancels, the hotel might lose money if they can’t rebook the room. One factor that might play a role is lead time—how far in advance the reservation was made. We explore whether people who book earlier are more or less likely to cancel. Understanding this connection can help hotels predict cancellations better and adjust their policies, such as flexible pricing or incentives for early bookings, benefiting both their business and their guests, making life easier for both travelers and hotel staff.

Both questions tackle challenges that many of us face when traveling. By analyzing the data, we hope to uncover insights that improve the experience for everyone involved, whether it’s choosing the most reliable airline or reducing the stress of last-minute cancellations.


## Data

We have two data set for analyzing our questions. The first one is the "Flights" dataset which contains 336770 observations, with each representing a single flight. Key variables used in the analysis include Airline Name, Scheduled Time, Actual Time, and Delay. Airline Name includes 12 most frequently seen U.S. airways, such as DL—Delta Air Lines, AS —Alaska Airlines, B6—JetBlue Airways, etc. Scheduled Time is the intended departure and arrival time, Actual Time is the actual departure and arrival time (in 24-hour). Delay is the difference between the actual and scheduled departure times of the flight, in minutes.

## Results
To effectively measure scheduling accuracy, ignoring the causations of how and why certain delays took place is a necessary approach since the overall core element is the final arrival time delay compared to scheduled ones. By focusing on the outcome of flights arrivals time accuracy, examination based on different airlines could potentially lead to determination of the overall performance. First Through separating delays to corresponding categories, all arrival delays of airlines are regrouped to conduct stacked bar charts along with boxplots and piecharts. 



Through examination it could be generalized that airlines with the majority of flights delayed less than 15 minutes, such as HA and AS, showcased high reliability in scheduling. Conversely, airlines like F9 and FL displayed a higher proportion of delays exceeding 45 minutes, which indicates potential major delays. This differentiation highlights the variability in performance and could help identify carriers excelling in time management versus those requiring operational improvements. By removing outliers, the boxplots focused on the typical delay experience. Airlines with consistently low delays, graphed as narrow interquartile ranges in figures like HA and AA, demonstrated strong scheduling accuracy. Wider ranges for carriers such as F9 and FL suggest potential inconsistencies in flight services, which could be a concern. Through emphasizing the percentage of flights in each delay category, it is found that airlines like HA, with over 85% of flights delayed less than 15 minutes, stood out positively. On the other hand, however, carriers with substantial delays in higher categories signal potential inefficiencies, depending on the expectations of the specific major demand.


Another concerning topic is the association between lead time and potential cancellation for hotels. To grasp a better understanding of the mass data, analysis of parameters such as mean, median, standard deviation regarding whether lead time is associated with cancellation or not appears crucial, as they would suggest possible correlation. Boxplot of cancellation status and lead time depicts clearer visualization for median and quantiles and excluding certain outliers, while violin plot regarding lead time distribution by cancellation status enables insights into the shape of distribution along with the mean, leading to generating potential correlation between cancellation status and lead time.

The violin plot demonstrates a visibly higher mean lead time for canceled bookings compared to non-canceled ones. For cancelled bookings, the distribution is skewed toward longer lead times, suggesting that bookings made further in advance have a higher likelihood of cancellation. Conversely, non-canceled bookings show a more concentrated distribution near zero lead time, indicating a preference for last-minute bookings.

The boxplot provides clear quantile-based insights. The median lead time for canceled bookings is noticeably higher than for non-cancelled bookings. Outliers in canceled bookings reflect certain extreme cases where bookings with comparably long lead times could eventually lead to being cancelled. For non-cancelled bookings, the interquartile range is narrower, signifying less variability in lead times, therefore suggesting that non-canceling customers often make bookings closer to their travel dates.
It could be summarized that long lead times typically indicate increased uncertainty and higher possibility of cancellation, while short lead times are generally associated with fewer cancellations, reflecting bookings of urgent demand.

Such insight could be intriguing for hotels seeking to mitigate revenue losses from cancellations, from which policies such as incentivizing early non-refundable bookings or flexible cancellations for last-minute bookings could be implemented to improve customer satisfaction and minimize cancellations to improve overall revenue.
